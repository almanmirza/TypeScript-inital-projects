```python
# Function to display basic dataset information
def display_dataset_info():
    while True:
        print("\n\n\t--- Dataset Information Menu ---")
        print("1. Display Head of the Dataset")
        print("2. Display Tail of the Dataset")
        print("3. Display Shape of the Dataset")
        print("4. Display Basic Information of the Dataset")
        print("5. Display Count of Null Values in Each Column")
        print("6. Display Count of Duplicate Rows")
        print("7. Back to Main Menu")

        choice = input("Enter your choice (1-7): ")
        print("\n")

        if choice == '1':
            print("\nHead of the dataset:\n", df.head())
        elif choice == '2':
            print("\nTail of the dataset:\n", df.tail())
        elif choice == '3':
            print("\nShape of the dataset:", df.shape)
        elif choice == '4':
            print("\nInfo about the dataset:\n", df.info())
        elif choice == '5':
            print("\nCount of null values in each column:\n", df.isnull().sum())
        elif choice == '6':
            print("\nCount of duplicate rows:", df.duplicated().sum())
        elif choice == '7':
            break
        else:
            print("Invalid choice. Please try again.")

def preprocess_data(df):
    df = df.drop_duplicates()
    le = LabelEncoder()
    categorical_features = ['released', 'writer', 'rating', 'name', 'genre', 'director', 'star', 'country', 'company']
    for feature in categorical_features:
        df[feature] = le.fit_transform(df[feature])
    features = df[['released', 'writer', 'rating', 'name', 'genre', 'director', 'star', 'country', 'company', 'runtime', 'score', 'budget', 'year', 'votes']]
    target = df['gross']

    # Standardize the features
    scaler = StandardScaler()
    features = scaler.fit_transform(features)

    return features, target

# Function to train and evaluate a Decision Tree model
def decision_tree_model(X_train, X_test, y_train, y_test):
    model = DecisionTreeRegressor()
    model.fit(X_train, y_train)
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_accuracy = r2_score(y_train, train_predictions)
    test_accuracy = r2_score(y_test, test_predictions)
    print('Decision Tree Regression')
    print(f'\nFinal Training Accuracy: {train_accuracy*100:.2f}%')
    print(f'Final Test Accuracy: {test_accuracy*100:.2f}%')
    plot_actual_vs_predicted(y_train, train_predictions, y_test, test_predictions, 'Decision Tree')

# Function to train and evaluate a Linear Regression model
def linear_regression_model(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_accuracy = r2_score(y_train, train_predictions)
    test_accuracy = r2_score(y_test, test_predictions)
    print('Linear Regression')
    print(f'\nFinal Training Accuracy: {train_accuracy*100:.2f}%')
    print(f'Final Test Accuracy: {test_accuracy*100:.2f}%')
    plot_actual_vs_predicted(y_train, train_predictions, y_test, test_predictions, 'Linear Regression')

# Function to train and evaluate a Logistic Regression model
def logistic_regression_model(X_train, X_test, y_train, y_test):
    model = LogisticRegression(max_iter=1000)  # Increased max_iter for better convergence
    model.fit(X_train, y_train)
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_accuracy = r2_score(y_train, train_predictions)
    test_accuracy = r2_score(y_test, test_predictions)
    print('Logistic Regression')
    print(f'\nFinal Training Accuracy: {train_accuracy*100:.2f}%')
    print(f'Final Test Accuracy: {test_accuracy*100:.2f}%')
    plot_actual_vs_predicted(y_train, train_predictions, y_test, test_predictions, 'Logistic Regression')

# Function to train and evaluate a Random Forest model
def random_forest_model(X_train, X_test, y_train, y_test):
    model = RandomForestRegressor(n_estimators=50)  # Reduced number of estimators for faster execution
    model.fit(X_train, y_train)
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_accuracy = r2_score(y_train, train_predictions)
    test_accuracy = r2_score(y_test, test_predictions)
    print('Random Forest Regression')
    print(f'\nFinal Training Accuracy: {train_accuracy*100:.2f}%')
    print(f'Final Test Accuracy: {test_accuracy*100:.2f}%')
    plot_actual_vs_predicted(y_train, train_predictions, y_test, test_predictions, 'Random Forest')

# Function to train and evaluate a Support Vector Machine model
def svm_model(X_train, X_test, y_train, y_test):
    model = SVR(kernel='linear', max_iter=1000)  # You can experiment with other kernels like 'rbf' or 'poly'
    model.fit(X_train, y_train)
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_accuracy = r2_score(y_train, train_predictions)
    test_accuracy = r2_score(y_test, test_predictions)
    print('Support Vector Machine Regression')
    print(f'\nFinal Training Accuracy: {train_accuracy*100:.2f}%')
    print(f'Final Test Accuracy: {test_accuracy*100:.2f}%')
    print("\nNote: Negative values from our SVM model suggest that it may not be suitable for this dataset\n")
    plot_actual_vs_predicted(y_train, train_predictions, y_test, test_predictions, 'Support Vector Machine')

# Function to plot actual vs predicted values
def plot_actual_vs_predicted(y_train, train_predictions, y_test, test_predictions, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(y_train, train_predictions, color='blue', label='Train')
    plt.scatter(y_test, test_predictions, color='red', label='Test')
    plt.title(f'{title}: Actual vs Predicted Values')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    plt.legend()
    plt.show()

# Function to plot number of movies by gross category
def plot_gross_category():
    bins = [0, 1_000_000, 10_000_000, 100_000_000, 1_000_000_000]
    labels = ['0-$1M', '$1M-$10M', '$10M-$100M', '$100M-$1B']
    df['gross_category'] = pd.cut(df['gross'], bins=bins, labels=labels, right=False)
    category_counts = df['gross_category'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    category_counts.plot(kind='bar', width=0.45, color=['#1f77b5', '#ff7f0e', '#2ca02c', '#d62728'])
    plt.title('Number of Movies by Gross Category')
    plt.xlabel('Gross Category')
    plt.ylabel('Number of Movies')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to plot movie distribution by country
def plot_country_distribution():
    country_counts = df['country'].value_counts()
    fig, ax = plt.subplots(figsize=(8, 8))  # Make the plot bigger
    ax.pie(country_counts, autopct='')  # Remove percentages
    plt.title('Movie Distribution by Country')
    ax.legend(title="Countries", labels=country_counts.index[:10], loc='upper left')
    plt.show()

# Function to evaluate models
def model_evaluation_menu():
    while True:

        print("\n\n\t--- Model Evaluation Menu ---")
        print("1. Train and Evaluate Decision Tree Model")
        print("2. Train and Evaluate Linear Regression Model")
        print("3. Train and Evaluate Logistic Regression Model")
        print("4. Train and Evaluate Random Forest Model")
        print("5. Train and Evaluate Support Vector Machine Model")
        print("6. Back to Main Menu")

        choice = input("Enter your choice (1-6): ")
        print("\n")

        if choice in ['1', '2', '3', '4', '5']:
            X, y = preprocess_data(df)
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

        if choice == '1':
            decision_tree_model(X_train, X_test, y_train, y_test)
        elif choice == '2':
            linear_regression_model(X_train, X_test, y_train, y_test)
        elif choice == '3':
            logistic_regression_model(X_train, X_test, y_train, y_test)
        elif choice == '4':
            random_forest_model(X_train, X_test, y_train, y_test)
        elif choice == '5':
            svm_model(X_train, X_test, y_train, y_test)
        elif choice == '6':
            break
        else:
            print("Invalid choice. Please try again.")

# Function to plot graphs
def plot_menu():
    while True:
        print("\n\n--- Plotting Menu ---")
        print("1. Plot Number of Movies by Gross Category")
        print("2. Plot Movie Distribution by Country")
        print("3. Back to Main Menu")
        print("\n")
        choice = input("Enter your choice (1-3): ")
        print("\n")

        if choice == '1':
            plot_gross_category()
        elif choice == '2':
            plot_country_distribution()
        elif choice == '3':
            break
        else:
            print("Invalid choice. Please try again.")

# Main function to run the menu
def main():
    while True:
        print("\n\t\t\t|------------------------------------------------------|\n")
        print("\t\t\t|------------------------------------------------------|\n")
        print("\t\t\t\t\t MOVIE REVENUE PREDICTION\n")
        print("\t\t\t|------------------------------------------------------|\n")
        print("\t\t\t\tAlman Ahmad          L1F22BSAI0004")
        print("\t\t\t\tEman Sarfraz         L1F22BSAI0034")
        print("\t\t\t|------------------------------------------------------|\n")
        print("\t\t\t|------------------------------------------------------|\n\n")

        print("\n\t--- Main Menu ---")
        print("1. Display Dataset Information")
        print("2. Model Evaluation")
        print("3. Plotting")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            display_dataset_info()
        elif choice == '2':
            model_evaluation_menu()
        elif choice == '3':
            plot_menu()
        elif choice == '4':
            print("Exiting the program. Have a nice day!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
